```python

# Task 2: Quantitative Financial Analysis
# Technical Indicators

# 1. Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os

sys.path.append(os.path.abspath(".."))

from src.technical_indicators import (
    calculate_sma,
    calculate_ema,
    calculate_rsi,
    calculate_macd,
)
# 2. Load Historical Stock Data
df = pd.read_csv("../data/raw/AAPL.csv")

# 3. Data Cleaning and Preprocessing

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.sort_values("Date")
df = df.dropna()

# 4. Technical Indicator Calculations

df["SMA_20"] = calculate_sma(df, "Close", 20)
df["EMA_20"] = calculate_ema(df, "Close", 20)
df["RSI_14"] = calculate_rsi(df, "Close", 14)
df["MACD"], df["Signal_Line"] = calculate_macd(df, "Close")

# 5. Daily Return Calculation
df["Daily_Return"] = df["Close"].pct_change() * 100

# 6. Visualization: SMA and EMA

plt.figure(figsize=(14, 6))
plt.plot(df["Date"], df["Close"], label="Close Price")
plt.plot(df["Date"], df["SMA_20"], label="20-Day SMA")
plt.plot(df["Date"], df["EMA_20"], label="20-Day EMA")
plt.title("AAPL Close Price with SMA and EMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 4))
plt.plot(df["Date"], df["RSI_14"], label="RSI 14")
plt.axhline(70, linestyle="--", label="Overbought")
plt.axhline(30, linestyle="--", label="Oversold")
plt.title("AAPL Relative Strength Index")
plt.xlabel("Date")
plt.ylabel("RSI")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 4))
plt.plot(df["Date"], df["MACD"], label="MACD")
plt.plot(df["Date"], df["Signal_Line"], label="Signal Line")
plt.title("AAPL MACD Indicator")
plt.xlabel("Date")
plt.ylabel("MACD")
plt.legend()
plt.tight_layout()
plt.show()

print("""
Interpretation:
The SMA and EMA smooth short-term stock price fluctuations and help identify the overall price trend.
The RSI helps identify overbought and oversold conditions.
The MACD helps analyze momentum changes and possible trend reversals.
These indicators provide useful quantitative evidence for understanding stock price behavior.
""")
```

```python

# Task 2: Quantitative Analysis
# Technical Indicators


import pandas as pd
import matplotlib.pyplot as plt

# 1. Load Dataset

df = pd.read_csv("../data/raw/AAPL.csv")

print("Dataset Shape:", df.shape)

display(df.head())

# 2. Data Cleaning

df["Date"] = pd.to_datetime(df["Date"])

df = df.sort_values("Date")

df = df.dropna()

print("\nMissing Values:")
display(df.isnull().sum())

# 3. Calculate Technical Indicator

# Simple Moving Average (SMA)

df["SMA_20"] = df["Close"].rolling(window=20).mean()

# 4. Visualization

plt.figure(figsize=(14, 6))

plt.plot(df["Date"], df["Close"], label="Close Price")

plt.plot(df["Date"], df["SMA_20"], label="20-Day SMA")

plt.title("AAPL Close Price with 20-Day SMA")

plt.xlabel("Date")

plt.ylabel("Price")

plt.legend()

plt.tight_layout()

plt.show()

# 5. Interpretation

print("""
Interpretation:

The 20-day Simple Moving Average (SMA) smooths short-term price fluctuations and helps identify the overall trend direction.

When the stock price remains above the SMA, it may indicate bullish momentum and upward market movement.

When the stock price falls below the SMA, it may indicate weakening momentum or possible downward movement.

The SMA can therefore help investors identify trend changes and support investment decision-making.
""")
```
